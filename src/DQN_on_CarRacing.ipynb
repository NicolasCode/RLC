# DQN on Car Racing

```python
from utils.performer import Performer
from utils.interpreters import gym_rgb_interpreter
from agents.deepQ import Atari32CNNasQ
from utils.plot_utils import PlotQApprox2D
import warnings
warnings.filterwarnings('ignore')
```

```python
NN_parameters = {
    'nA':5,
    'alpha':1e-3,
    'optimizer_class':'Adam'
}
value_approximator = Atari32CNNasQ(parameters=NN_parameters)
agent_parameters = {
    'nA':5,
    'gamma':0.8,
    'epsilon':0.6,
    'NN':value_approximator,
    'target_network_latency':1,
    'len_exp':32,
    'batch_size':4,
    'num_epochs':1,
}
# env_parameters = {
#    'lap_complete_percent':0.02,
#    'continuous':False
#}
env_parameters = {
    'lap_complete_percent':0.08,
    'initial_round':80
}
state_interpreter = gym_rgb_interpreter()
perf = Performer(
    env_name='SpecialCarRacing',
    env_parameters=env_parameters,
    state_interpreter=state_interpreter,
    agent_name='agentsNN.DQN',
    agent_parameters=agent_parameters
)
```

```python
perf.run(num_rounds=120, sleep_time=0, from_file=True)
```

```python
perf.train(num_rounds=150, num_episodes=10, from_file=True)
```
