```python
import numpy as np
from agents.deepQ import MLP_as_Q
from utils.plot_utils import PlotQApprox2D
from agents.agentsNN import ExperienceDataset
import torch
from torch.distributions import MultivariateNormal
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
from prettytable import PrettyTable
```

Creamos un Perceptrón Multi Capa (MLP) para aproximar el valor de una pareja estado, acción mediante un objeto `MLP_as_Q` del módulo `agents.deepQ`, el cual instanciaremos con el nombre de `value_approximator`:

```python
MLP_parameters = {'sizes':[2, 20, 10, 3],
                  'intermediate_activation_function':torch.nn.Sigmoid(),
                  'last_activation_function':None,
                  'alpha':1e-3}
value_approximator = MLP_as_Q(parameters=MLP_parameters)
```

```python
def summary(model):
    table = PrettyTable(['Modules', 'Parameters'])
    total_params = 0
    for name, parameter in model.named_parameters():
        if not parameter.requires_grad: continue
        params = parameter.numel()
        table.add_row([name, params])
        total_params+=params
    print(table)
    print(f'Total Trainable Params: {total_params}')

summary(value_approximator.NN.model)
```

Observe que `value_approximator` tiene un método `predict()`, el cual devuelve el valor predecido de una pareja estado, acción. Como recién inicializamos el objeto, el valor predecido es aleatorio:

```python
state = torch.tensor([-.5, .01])
action = 2
state_action_value = value_approximator.predict(state=state,
                                                action=action)
print('Predicted value:', state_action_value)
```

Observe que predecimos $q(s, a)$ para cada estado $s$. La acción que el agente debe tomar (módulo $\epsilon$) es la que tenga mayor valor $q$. Los mejores valores y su correspondiente acción los podemos dibujar de la siguiente manera:

```python
dict_acciones = {0:"⬅", 1:"N", 2:"➡"}
state_scales = [[-1.2, 0.6], [-0.07, 0.07]]
q_plot = PlotQApprox2D(Q=value_approximator,
                       state_scales=state_scales, 
                       dict_acciones=dict_acciones,
                       mesh_size=7,
                       use_tensors=True)
q_plot.best_actions_with_values(max_deci=2)
```

Podemos decirle a la función que aprenda que la acción 2 tiene el valor de una distribución normal bivariada centrada en $(0,0)$. Esto modificará la función $q$ en muchos estados:

```python
def simulate_data(num_sample):
    # get scales
    x_position_scale = state_scales[0]
    x_velocity_scale = state_scales[1]
    # Create bookkeeping lists
    states = list()
    actions = [2] * num_sample
    updates = list()
    # Simulate total rewards
    for i in range(num_sample):
        x_position = np.random.uniform(low=x_position_scale[0], high=x_position_scale[1])
        x_velocity = np.random.uniform(low=x_velocity_scale[0], high=x_velocity_scale[1])
        state = torch.tensor([x_position, x_velocity], dtype=torch.float32)
        states.append(state)
        m = MultivariateNormal(torch.tensor([-0.2,0]), torch.tensor([[0.1, 0], [0, 0.01]]))
        logprob = m.log_prob(state).item()
        total_reward = np.exp(logprob)
        updates.append(total_reward)
    return states, actions, updates

num_sample = 8
states, actions, updates = simulate_data(num_sample=num_sample)
ds = ExperienceDataset(states, actions, updates)
ds_loader = DataLoader(ds, batch_size=4, shuffle=True)
```

```python
counter = 0
for batch_states, batch_actions, batch_updates in ds_loader:
    counter += 1
    print('-'*10 + f'Batch number {counter}' + '-'*10)
    print(f'states:\n\t{batch_states}')
    print(f'actions:\n\t{batch_actions}')
    print(f'updates:\n\t{batch_updates}')
    print('\n')
```

```python
num_sample = 1000
states, actions, updates = simulate_data(num_sample=num_sample)
ds = ExperienceDataset(states, actions, updates)
ds_loader = DataLoader(ds, batch_size=64, shuffle=True)
```

```python
num_epochs = 200
value_approximator.reset()
for e in tqdm(range(num_epochs)):
    value_approximator.learn(ds_loader)
```

```python
plt.plot(value_approximator.losses)
```

```python
# Actualize q_plot
q_plot.Q = value_approximator
# Plot best action with its corresponding value
q_plot.best_actions_with_values(max_deci=1)
```
