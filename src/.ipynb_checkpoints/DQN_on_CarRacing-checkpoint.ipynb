# DQN on Car Racing

```python
from utils.performer import Performer
from utils.interpreters import gym_rgb_interpreter
from agents.deepQ import Atari32CNNasQ
from utils.plot_utils import PlotQApprox2D
import warnings
warnings.filterwarnings('ignore')
```

```python
NN_parameters = {'nA':5,
                  'alpha':1e-3,
                  'optimizer_class':'Adam'
                 }
value_approximator = Atari32CNNasQ(parameters=NN_parameters)
agent_parameters = {'nA':5,
                    'gamma':0.8,
                    'epsilon':0.6,
                    'NN':value_approximator,
                    'target_network_latency':1,
                    'len_exp':8,
                    'batch_size':4,
                    'num_epochs':1,
                    }
env_parameters = {continuous=False}
# env_parameters = {'lap_complete_percent':0.08,
#                  'initial_round':90}
state_interpreter = gym_rgb_interpreter()
perf = Performer(env_name='CarRacing-v2',
                 env_parameters=env_parameters,
                 state_interpreter=state_interpreter,
                 agent_name='agentsNN.DQN',
                 agent_parameters=agent_parameters)
```

```python
perf.run(num_rounds=120, sleep_time=0, from_file=False)
```

```python
perf.train(num_rounds=80, num_episodes=1, from_file=False)
```

```python
perf.agent.dones
```

```python
perf.test(num_rounds=80, num_episodes=10)
```

```python
perf.run(from_file=True, no_exploration=True, sleep_time=0, num_rounds=80)
```
