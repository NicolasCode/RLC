# Ejemplo de un agente Q-learning aprendiendo a superar el Blackjack

```python
from utils.performer import Performer
from utils.interpreters import blackjack_interpreter
```

En primer lugar, creamos los parámetros del agente y luego creamos el objeto `Performer`. Esta clase contiene los métodos que vamos a utilizar para que los agentes resuelvan los entornos.

```python
agent_parameters = {'nA':2,
                    'nS':32*11*2,
                    'gamma':0.8,
                    'epsilon':0.05,
                    'alpha':1e-2}
perf = Performer(env_name='Blackjack-v1',
                 env_parameters={},
                 state_interpreter=blackjack_interpreter,
                 agent_name='Q_learning',
                 agent_parameters=agent_parameters)
```

```python
#perf.run(from_file=False, to_video=True)
```

```python
perf.sweep(parameter='alpha',
           values=[0.01, 0.1, 0.2],
           num_episodes=1000,
           num_simulations=100)
```

```python
perf.sweep(parameter='epsilon',
           values=[0, 0.01, 0.1],
           num_episodes=1000,
           num_simulations=100)
```

```python
perf.train(num_episodes=1000)
```

```python
perf.test(num_episodes=1000)
```

```python
perf.run(from_file=True, no_exploration=True, to_video=True)
```
