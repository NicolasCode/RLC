# Using linear features to approximate policy and V

```python
from utils.performer import Performer
from utils.interpreters import gym_interpreter1
from agents.linearPolicy import LinearPolicy, LinearV
from utils.plot_utils import PlotQApprox2D, PlotGridValues
from torch.distributions import MultivariateNormal
import seaborn as sns
import torch
```

## Using linear features to predict policy

```python
policy_parameters = {'nA':3,
                     'nS':2,
                     'alpha':1e-5,
                     'beta':1,
                     'gamma':0.99
                     }
policy = LinearPolicy(parameters=policy_parameters)
```

```python
state = [-.5, .01]
action_probabilities = policy.predict(state=state)
print('Probabilities for actions:', action_probabilities)
```

## Using linear features to approximate state value

```python
value_fn_parameters = {'nS':2,
                       'alpha':1e-3,
                       'beta':1,
                       'gamma':0.99
                       }
value_approximator = LinearV(parameters=value_fn_parameters)
```

```python
state_value = value_approximator.predict(state=state)
print('Predicted state value:', state_value)
```

```python
def update_functions(state, action, next_state, reward, gamma):
    # Determine delta
    G = reward + gamma * value_approximator.predict(next_state)
    delta = G - value_approximator.predict(state)
    # Update policy
    policy.learn(state, action, delta)
    # Update value estimate
    value_approximator.learn(state, delta)
```

```python
state = [-.5, .01]
next_state = [-.55, 0]
reward = -1
action = 1
gamma = 0.99
update_functions(state, action, next_state, reward, gamma)
```

```python
dict_acciones = {0:"⬅", 1:"N", 2:"➡"}
state_scales = [[-1.2, 0.6], [-0.07, 0.07]]
q_plot = PlotQApprox2D(Q=policy,
                       state_scales=state_scales, 
                       dict_acciones=dict_acciones,
                       mesh_size=8)
X, Y = q_plot.create_mesh()
print(X, Y)
```

Observe que predecimos $q(s, a)$ para cada estado $s$. La acción que el agente debe tomar (módulo $\epsilon$) es la que tenga mayor valor $q$. Los mejores valores y su correspondiente acción los podemos dibujar de la siguiente manera:

```python
policy_best = []
V = [] 
pp = PlotGridValues(shape=(10,10), dict_acciones=dict_acciones)
pp.plot_policy_and_values(policy_best, V)
```

Podemos decirle a la función que aprenda que la acción 2 tiene el valor de una distribución normal bivariada centrada en $(0,0)$. En la siguiente celda vemos una visualización de esta función:

```python
# Create mesh from bidimensional states
mesh_size = 7
x_lims = state_scales[0]
X = np.linspace(x_lims[0], x_lims[1], mesh_size)
y_lims = state_scales[1]
Y = np.linspace(y_lims[0], y_lims[1], mesh_size)
# Create labels
xticklabels = [str(round(x, 1)) for x in X]
yticklabels = [str(round(y, 1)) for y in Y]
# Create q table from Q function
actions = list(dict_acciones.keys())
n = len(actions)
m = MultivariateNormal(torch.tensor([-0.2,0]), torch.tensor([[0.1, 0], [0, 0.001]]))
q_table = np.zeros((mesh_size, mesh_size, n))
for j in range(mesh_size):
    for i in range(mesh_size):
        state = (X[j], Y[i])
        state = torch.tensor(state, dtype=torch.float32)
        for a in range(n):
            q_table[(mesh_size - 1) - i, j, a] = np.exp(m.log_prob(state).item())
V = np.max(q_table, axis=2)
sns.heatmap(V,
            xticklabels=xticklabels,
            yticklabels=yticklabels)
```

En la siguiente celda entrenamos al `value_approximator` sobre una muestra aleatoria de datos, consistente un valor simulado para cada pareja (estado, acción=2):

```python
# get scales
x_position_scale = state_scales[0]
x_velocity_scale = state_scales[1]
# Simulate total rewards
for i in range(1000):
    x_position = np.random.uniform(low=x_position_scale[0], high=x_position_scale[1])
    x_velocity = np.random.uniform(low=x_velocity_scale[0], high=x_velocity_scale[1])
    action = 2
    m = MultivariateNormal(torch.tensor([-0.2,0]), torch.tensor([[0.1, 0], [0, 0.001]]))
    logprob = m.log_prob(torch.tensor([x_position, x_velocity])).item()
    total_reward = np.exp(logprob)
    value_approximator.learn(state=(x_position, x_velocity), 
                             action=action, 
                             update=total_reward)
# Actualize q_plot
q_plot.Q = value_approximator
# Plot best action with its corresponding value
q_plot.best_actions_with_values(max_deci=1)
```
